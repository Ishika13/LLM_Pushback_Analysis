```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# Assume df_1, df_2, df_3, df_4 are already defined
dfs = [(df_1, "Experiment 1"),
       (df_2, "Experiment 2"),
       (df_3, "Experiment 3"),
       (df_4, "Experiment 4")]

all_dataframes = []
for df, exp_name in dfs:
    df["Experiment"] = exp_name
    all_dataframes.append(df)

# Merge all experiments into one DataFrame
combined_df = pd.concat(all_dataframes, ignore_index=True)

# Rename columns for easier analysis
combined_df.rename(columns={
    "meta-llama/Llama-3.1-8B-Instruct_predictions_error_acknowledgement": "llama_ack",
    "meta-llama/Llama-3.1-8B-Instruct_predictions_error_correction": "llama_correct",
    "google/gemma-2-2b-it_predictions_error_acknowledgement": "gemma_ack",
    "google/gemma-2-2b-it_predictions_error_correction": "gemma_correct"
}, inplace=True)

# Replace extra whitespace in boolean strings
combined_df.replace({"False \n": "False", "True \n": "True"}, inplace=True)

# Define conversion function for booleans
def convert_to_bool(val):
    if isinstance(val, bool):
        return val
    elif isinstance(val, str):
        return True if val.strip() == "True" else False
    return False

# Apply conversion to the relevant columns
cols_to_convert = ["llama_ack", "llama_correct", "gemma_ack", "gemma_correct"]
for col in cols_to_convert:
    combined_df[col] = combined_df[col].apply(convert_to_bool)

# -------------------------
# Aggregated Metrics Per Experiment
# -------------------------
experiment_means = combined_df.groupby("Experiment").agg({
    "llama_ack": "mean",
    "gemma_ack": "mean",
    "llama_correct": "mean",
    "gemma_correct": "mean"
}).reset_index()

print("Mean Performance Metrics per Experiment:")
display(experiment_means)

# Save the aggregated table to CSV
experiment_means.to_csv("experiment_means.csv", index=False)

# -------------------------
# Line Graphs for Error Acknowledgment and Correction Rates
# -------------------------
# Acknowledgment Rates
plt.figure(figsize=(10,6))
plt.plot(experiment_means['Experiment'], experiment_means['llama_ack'], marker='o', label='Llama Acknowledgment')
plt.plot(experiment_means['Experiment'], experiment_means['gemma_ack'], marker='o', label='Gemma Acknowledgment')
plt.title("Error Acknowledgment Rates Across Experiments")
plt.xlabel("Experiment")
plt.ylabel("Acknowledgment Rate")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
# Save the figure to a file
plt.savefig("acknowledgment_rates.png")
plt.show()

# Correction Rates
plt.figure(figsize=(10,6))
plt.plot(experiment_means['Experiment'], experiment_means['llama_correct'], marker='o', label='Llama Correction')
plt.plot(experiment_means['Experiment'], experiment_means['gemma_correct'], marker='o', label='Gemma Correction')
plt.title("Error Correction Rates Across Experiments")
plt.xlabel("Experiment")
plt.ylabel("Correction Rate")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("correction_rates.png")
plt.show()

# -------------------------
# Breakdown by Error Type Across Experiments
# -------------------------
error_type_summary = combined_df.groupby(["Experiment", "Error Type"]).agg({
    "llama_ack": "mean",
    "gemma_ack": "mean",
    "llama_correct": "mean",
    "gemma_correct": "mean"
}).reset_index()

print("Performance by Error Type per Experiment:")
display(error_type_summary)

# Save this summary as a CSV file
error_type_summary.to_csv("error_type_summary.csv", index=False)

plt.figure(figsize=(12,8))
sns.barplot(data=error_type_summary, x="Error Type", y="llama_ack", hue="Experiment")
plt.title("Llama Acknowledgment by Error Type Across Experiments")
plt.xticks(rotation=45)
plt.legend(title="Experiment")
plt.tight_layout()
plt.savefig("llama_ack_by_error_type.png")
plt.show()

plt.figure(figsize=(12,8))
sns.barplot(data=error_type_summary, x="Error Type", y="llama_correct", hue="Experiment")
plt.title("Llama Correction by Error Type Across Experiments")
plt.xticks(rotation=45)
plt.legend(title="Experiment")
plt.tight_layout()
plt.savefig("llama_correct_by_error_type.png")
plt.show()

# -------------------------
# Scatter Plots: Acknowledgment vs Correction
# -------------------------
# For Llama Model
plt.figure(figsize=(10,6))
sns.scatterplot(data=experiment_means, x="llama_ack", y="llama_correct", hue="Experiment", s=100)
plt.title("Llama: Acknowledgment vs Correction Rates by Experiment")
plt.xlabel("Acknowledgment Rate")
plt.ylabel("Correction Rate")
plt.legend(title="Experiment")
plt.grid(True)
plt.tight_layout()
plt.savefig("llama_ack_vs_correct.png")
plt.show()

# For Gemma Model
plt.figure(figsize=(10,6))
sns.scatterplot(data=experiment_means, x="gemma_ack", y="gemma_correct", hue="Experiment", s=100)
plt.title("Gemma: Acknowledgment vs Correction Rates by Experiment")
plt.xlabel("Acknowledgment Rate")
plt.ylabel("Correction Rate")
plt.legend(title="Experiment")
plt.grid(True)
plt.tight_layout()
plt.savefig("gemma_ack_vs_correct.png")
plt.show()

# -------------------------
# Combined Model Performance
# -------------------------
experiment_means['combined_ack'] = (experiment_means['llama_ack'] + experiment_means['gemma_ack']) / 2
experiment_means['combined_correct'] = (experiment_means['llama_correct'] + experiment_means['gemma_correct']) / 2

plt.figure(figsize=(10,6))
plt.plot(experiment_means['Experiment'], experiment_means['combined_ack'], marker='o', label='Combined Acknowledgment')
plt.plot(experiment_means['Experiment'], experiment_means['combined_correct'], marker='o', label='Combined Correction')
plt.title("Combined Error Metrics Across Experiments")
plt.xlabel("Experiment")
plt.ylabel("Rate")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("combined_metrics.png")
plt.show()
```
